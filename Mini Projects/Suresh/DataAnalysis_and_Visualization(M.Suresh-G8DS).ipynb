# Number of Athletes in each country

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.countplot(x='NOC', data=athletes[:20], palette='viridis')

plt.title('Total Number of Athletes in Each NOC')
plt.xlabel('NOC')
plt.ylabel('Number of Athletes in 100s')

# Adding a legend


plt.show()


```

<h2><font color="blue">Number of athletes in each discipline</font></h2>

```python
plt.figure(figsize=(10,10))
athletes['Discipline'].value_counts().plot(kind="bar",color="g")
plt.title("Number of athletes in each discipline")
plt.show()

```

<h2><font color="green">Distribution of Disciplines in particular counter</font></h2>

```python
india_data =athletes[athletes['NOC'] == 'India']
discipline_counts = india_data['Discipline'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
discipline_counts.plot(kind='barh', color='r')
plt.title('Distribution of Disciplines in India',weight="bold")
plt.ylabel('Discipline')
plt.xlabel('Count')
plt.show()
```

# Total medals for each country

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'medals' is your DataFrame with the given columns
# Replace 'Team/NOC' and 'Total' with your actual column names

plt.figure(figsize=(20,10))
plt.bar(x=medals['Team/NOC'],height=medals['Total'], color='purple')
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.title('Total Medals for Each Country')
plt.xticks(rotation=90) 
plt.show()
```

<h2 style color="green">Top 10 countries with Medals</h2>

```python
A = plt.pie(medals['Total'][:10], 
        labels=medals['Team/NOC'][:10],
        startangle=90,
        autopct='%1.1f%%')
plt.title("Pie Chart of Top 10 Countries with Medals",weight="bold")
plt.show()
```

# Number of medals a particular country won

```python
country=medals[medals['Team/NOC'] == 'India']

gold=country['Gold'].sum()
silver=country['Silver'].sum()
bronze=country['Bronze'].sum()

medal_counts = [gold, silver, bronze]
medal_labels = ['Gold', 'Silver', 'Bronze']


plt.figure(figsize=(4,4))
plt.pie(medal_counts,labels=medal_labels, autopct='%1.1f%%')
plt.legend(loc="upper right")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))


sns.distplot(medals['Gold'],color='red', label='Gold')

sns.distplot(medals['Silver'],color='blue', label='Silver')

sns.distplot(medals['Bronze'],color='brown', label='Bronze')

plt.title('Distribution of Gold,Silver and Bronze Medals')
plt.xlabel('Number of Medals')
plt.ylabel('Density')
plt.legend()
plt.show()



```

# Pairplot

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(1,4))


subset = medals[['Gold', 'Silver', 'Bronze']]
sns.pairplot(subset, palette='viridis')
plt.suptitle('Pair Plot of Gold, Silver, Bronze Medals', y=1.02)
plt.show()

```
