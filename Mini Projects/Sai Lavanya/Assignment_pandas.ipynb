```python
data['condition']=data['condition'].fillna(data['condition'].mean())
data['odometer']=data['odometer'].fillna(data['odometer'].mean())
data['mmr']=data['mmr'].fillna(data['mmr'].mean())
data['sellingprice']=data['sellingprice'].fillna(data['sellingprice'].mean())
```

```python
data['make']=data['make'].fillna(data['make'].mode().iloc[0])
data['model']=data['model'].fillna(data['model'].mode().iloc[0])
data['trim']=data['trim'].fillna(data['trim'].mode().iloc[0])
data['body']=data['body'].fillna(data['body'].mode().iloc[0])
data['transmission']=data['transmission'].fillna(data['transmission'].mode().iloc[0])
data['vin']=data['vin'].fillna(data['vin'].mode().iloc[0])
data['state']=data['state'].fillna(data['state'].mode())
data['color']=data['color'].fillna(data['color'].mode().iloc[0])
data['interior']=data['interior'].fillna(data['interior'].mode().iloc[0])
data['seller']=data['seller'].fillna(data['seller'].mode())
data['saledate']=data['saledate'].fillna(data['saledate'].mode().iloc[0])
```
